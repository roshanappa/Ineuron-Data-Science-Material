#### Bivariant Graph Analysis -Two varaibles are used.

```python
sns.lmplot(x='total_bill',y='tip',data=df1)  #lm-linear model
plt.show()
```

```python
sns.lmplot(x='total_bill',y='tip',data=df1,fit_reg=False)  #lm-linear model
plt.show()
```

#### Multivariate Graph analysis - more than two variables are used.

```python
sns.lmplot(x='total_bill',y='tip',data=df1,fit_reg=False,hue='sex')  #add hue
plt.show()
```

```python
sns.lmplot(x='total_bill',y='tip',data=df1,fit_reg=False,hue='smoker')  #remove the linear line
plt.show()
```

```python
sns.lmplot(x='total_bill',y='tip',data=df1,fit_reg=False,hue='smoker',markers=['x','o'])  #add markers
plt.show()
```

```python
sns.lmplot(x='total_bill',y='tip',data=df1,fit_reg=False,hue='smoker',markers=['x','o'],palette='winter')  #add pallete
plt.show()
```

```python
sns.lmplot(x='total_bill',y='tip',data=df1,fit_reg=False,row='day')  #lm-linear model
plt.show()
```

```python
sns.lmplot(x='total_bill',y='tip',data=df1,fit_reg=False,col='time')  #lm-linear model
plt.show()
```

```python
sns.lmplot(x='total_bill',y='tip',data=df1,fit_reg=False,col='time',row='smoker')  #lm-linear model
plt.show()
```

```python
sns.lmplot(x='total_bill',y='tip',data=df1,fit_reg=False,col='time',row='smoker',hue='sex')  #lm-linear model
plt.show()
```

```python
sns.lmplot(x='total_bill',y='tip',data=df1,fit_reg=False,col='day',row='smoker',hue='sex')  #lm-linear model
plt.show()
```
