## MnagoDB Database 

- MongoDB, developed by MongoDB Inc..
- It is a cross-platform database program. It is categorized as a NoSQL database program which stores data in JSON-like documents.
- It is categorized as a `NoSQL` database program which stores data in `JSON-like` documents.
- A document is a set of key-value pairs defined by a schema which defines how the data is structured and organized in a database. A group of documents makes up a collection, and documents within the same collection can have different schema. This is referred to as dynamic schema.

- MongoDB, a distributed database at its core, is known to provide high performance, scalability, and flexibility in querying and indexing data. It is easy to integrate with applications as the document model can map to objects in the code. It provides drivers for major languages and there is also a lot of professional support available.

## Difference Between NO-SQL and MySQL Database.

- **SQL** databases are primarily referred to as **Relational Databases (RDBMS).**

- **NoSQL** databases are primarily referred to as **Non-relational or Distributed Database.**

## Now we will see the methods which are avaliable inside the MongoDB.

### 1. Find Method.

```python
find_first_method = collection.instance1.find_one() #Get a single document from the database.
find_first_method
```

```python
collection.instance1.find() #Get a single document from the database.
```

```python
for i in collection.instance1.find(): #it gives me all the records avalible inside the database.
    print(i)
    print(i)
```

### 2.Query and filter out data in MongoDB.

```python
query = {'Fees': 12000}
result = collection.instance1.find(query)

for i in result:
    print(i)
```

### Link :- https://www.mongodb.com/blog/post/getting-started-with-python-and-mongodb

```python
query = {'Fees': {"$gt":1000,"$lt":2000}} #gt is nothing but the greater than.It is initiated with the help of `$` sign.
result = collection.instance1.find(query) #lt is nothing but the lesser than.
                                          #gte = greater than equal to
                                          #lte = less than equal to

for i in result:
    print(i)
```

### 3. Delete one or Many Documents In MongoDB.
