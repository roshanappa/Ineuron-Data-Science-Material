## Scalling In Machine Learning.

- In the machine learning we have lot of feature in the different ranges and magnitude.
- While training the model we required the feature in the numerical format.
- Each column having their own behaviour that is why they have diffrent magnitude.
- In the machine learning we required distance measurement from base line espcially whenver we do have parametrc approach in the particular algorithm like KNN,Linear Regression,Logistic Regression etc.
- This meansurement should be so far for regression because model will get overfitted because of high variance in other word data points will be avaliable far away from its mean and alorithm takes lot of training time because it usually takes time to cover each point in order get train.

Meanwhile some algorithm gives higher importance to higher magnitude even it is not important as far as our business decision concern like some 1000 gram is equal 1 kg.
In this case algorithm gives high important to 1000 gram not to the 1 Kg.
In order to avoid this kind of problem we really required scalling.
