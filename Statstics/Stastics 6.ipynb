## Biased Enstimation.

- Inside the bias estimation we are taking denominator value as **`n`**.
- if you have sample which is avaliable inside the population in dense form.
- The possibility of sample standard Deviation will be nearer the sample center and that will be far from the population Standard Deviation.
- if you calculate the standard deviation with denominator value as **`n`**,then your std about sample will not get any approximation with standard deviation of the  population  and this will perform as  biased estimator.

- Becuase if we use **`n`** as denominator the value will become lesser and maginitude of standard deviation of sample will become samller.

If we can see that  biased estimator is going to work on the sample,but at the certain data points wheere we are founding bounch of data points at time and at that cluster if we are making the sample and calculate the standrad dviation of that cluster that will be far away from the actual population standard deviation.

## UnBiased Estimation.

- Inside the bias estimation we are taking denominator value as **`n-1`**.
- try to calculate the standard deviation of sample by keeping denominator **`n-1`*.
- Then that sample standard deviation will try to match with population standard deviation. 
- If you calculate the standard deviation with denominator value as **`n-1`**,then your std about sample will get approximation with standard deviation of the  population  and this will perform as  unbiased estimator.

If you see mathematically,the division of any value with **`n-1`** ,the output will be in higher maginitude and that means the whatever the value of Standard deviation came from the formualae of unbiased estimator that will get approximate with standard deviation of the population.

- This is again question that we are not considering the **`n-2,n-3,n-4...n-m`** for calculating the standard deviation of sample form poupulation.

- Because Scientist have done experiement of the **`n-m`** and they have found good result only at the **`n-1`** denominator value that is why its been choose.
- As the value of **`m`** is getting increase inside **`n-m`** the value of standard deviation of sample is also moving far away from population standard deviation and vice versa.
- So appropriate value is **`n-1`** for calculating the population sampling standard deviation calculation.
