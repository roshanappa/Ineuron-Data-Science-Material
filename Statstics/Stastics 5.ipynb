## 4.Power Law.

If you see the 20% people are wealthy and 80% are poor called as Parreto distribution.

https://en.wikipedia.org/wiki/Pareto_distribution

Lets suppose we have log normal distribution and we want to convert that into the normal distribution by usng log 

Logn ---> Log 

In order to make the Power Law distribution or parreto distribution in the form of Gaussian Distribution,it can done by using the **`Box-Cox Transformation`**

In order to check wheather it is become gaussian distributed or not we have method called **`Q-Q Plot`.**

### Mathematics behind it

For making parreto distribution to Gaussian distribution we will have to value of aplha.

- If the alpha is not equal to zero then it will use Box-Cox transformation by using below formuale and that is going to help to make Gaussian Distribution.
- If alpha is equal to zero then it will use log transformation of that data in order to convert the gaussian distribution.

Alpha is height decider parameter 

Most of the machine learning assume the data should be in gaussian distribution for that we want to transform the data into gaussian distribution. 

Then we will able make some good assumptions on top of that data prediction.

#### `Q-Q plot(Quantile Quantile Plot)`

```python
import numpy as np
import pandas as pd
import statsmodels.api as sm
import pylab as py
```

```python
data_points = np.random.normal(0,1,100) #taking 100 values where the mean = 0 and std = 1
data_points
```

```python
sm.qqplot(data_points,line="45")
py.show()
```

The datapoint to getting surrounded the ideal line of standard normal distribution.

```python
import seaborn as sns
sns.displot(data_points,kind="kde")
```
