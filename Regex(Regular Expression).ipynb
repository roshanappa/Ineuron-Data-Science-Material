### Group,Start,end and span

```python
import re
test_string = '123abc456789abc123ABC'
pattern = re.compile(r'123')
matches = pattern.finditer(test_string)

for match in matches:
    print(match)
```

```python
import re
test_string = '123abc456789abc123ABC'
pattern = re.compile(r'123')
matches = pattern.finditer(test_string)

for match in matches:
    print(match.span())
    print(match.start())
    print(match.end())
    print(match.group())
    print()
```

```python
import re
test_strings = '123abc456789abc123ABC'
pattern = re.compile(r'abc')
matche = pattern.finditer(test_strings)

for match in matche:
    print(match.group(0))
```

It is use for the every charecter exacept new line.

### Quantifiers

```python
import re # * it is going to work for the o to more digit 
text = r"file_124774"
pattern = re.compile(r"\d*")
matches = pattern.finditer(text)

for match in matches:
    print(match)
```

```python
import re # * it is going to work for the 1 to more digit after the "_"
text = r"file_124774"
pattern = re.compile(r"\d+")
matches = pattern.finditer(text)

for match in matches:
    print(match)
```

```python
import re # + it is going to work for the 1 to more digit after the "_"
text = r"file_124774"
pattern = re.compile(r"_\d+")
matches = pattern.finditer(text)

for match in matches:
    print(match)
```

```python
import re # ? is going to work for exceptional hadeling there might be anything which we dont know
text = r"file_!124774"
pattern = re.compile(r"_?\d+")
matches = pattern.finditer(text)

for match in matches:
    print(match)
```

```python
import re # {} it is used to get the number digit by providing extact one number inside the curly bracket.
text = r"file_124774"
pattern = re.compile(r"\d{4}") #4 provided so will get only four extact number from left to right
matches = pattern.finditer(text)

for match in matches:
    print(match)
```

```python
import re # {d,d} it is used to get the digit by providing number range inside the curly bracket.
text = r"file_124774"
pattern = re.compile(r"\d{1,10}")
matches = pattern.finditer(text)

for match in matches:
    print(match)
```

### Conditions

Either or conditions

```python
my_string = """

hello world 
12455
2020-05-20
Mr Simpson
Mrs Simpson
Mr. Brown
Ms Smith
Mr. T
"""
```

```python
import re
pattern = re.compile(r'Mr')
matches   = pattern.finditer(my_string)

for match in matches:
    print(match)
```

```python
import re
pattern = re.compile(r'Mr\s\w+')
matches   = pattern.finditer(my_string)

for match in matches:
    print(match)
```

```python
import re
pattern = re.compile(r'Mr\.\s\w+')
matches   = pattern.finditer(my_string)

for match in matches:
    print(match)
```

```python
import re
pattern = re.compile(r'Mr\.?\s\w+') #made (.) optional by using ? mark.
matches   = pattern.finditer(my_string)

for match in matches:
    print(match)
```

```python
import re #Here I am using the | to get the mutiple output 
pattern = re.compile(r'(Mr|Mrs|Ms)\.?\s\w+') #made (.) optional by using ? mark.
matches   = pattern.finditer(my_string)

for match in matches:
    print(match)
```

### Grouping.

```python
import re
emails = """
pythonengineer@gmail.com
Python-engineer@gmx.de
python-engineer123@my-domain.org
"""
pattern = re.compile(r'[A-Za-z0-9-]+\@[a-z-]+\.(com|de|org)')

matches = pattern.finditer(emails)

for match in matches:
    print(match)
```

```python
emails = """
pythonengineer@gmail.com
Python-engineer@gmx.de
python-engineer123@my-domain.org
"""
pattern = re.compile(r'[A-Za-z0-9-]+\@[a-z-]+\.[a-z-]+')

matches = pattern.finditer(emails)

for match in matches:
    print(match)
```

```python
emails = """
pythonengineer@gmail.com
Python-engineer@gmx.de
python-engineer123@my-domain.org
"""
pattern = re.compile(r'([A-Za-z0-9-]+)\@([a-z-]+)\.([a-z-]+)')

matches = pattern.finditer(emails)

for match in matches:
    print(match.group(0))
    print(match.group(1))
    print(match.group(2))
    print(match.group(3))
    print()
```

### Replacement of substring with group

```python
urls = """
http://python-engineer.com
https://www.python-engineer.org
http://www.pyeng.net

"""

pattern = re.compile(r'(https?\:)//(www.)?([a-z-]+)\.([a-z]+)')
matches = pattern.finditer(urls)

sub_url = pattern.sub(r'\3.\4',urls)
print(sub_url)
```
